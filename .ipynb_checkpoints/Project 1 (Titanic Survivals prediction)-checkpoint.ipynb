# Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Import Dataset

```python
data=pd.read_csv('train.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

# Data Preprocessing

## Dealing with missing value

```python
# to know missing value
data.isna().sum()
```

```python
# to show in diagram missing value
sns.heatmap(data.isna())
```

```python
#to solve missing value in column "age" by using average value of age 
data['Age'] =data['Age'].fillna(data['Age'].mean())
```

```python
sns.heatmap(data.isna())
```

```python
data.drop('Cabin',axis=1,inplace=True)
```

```python
data.head()
```

```python
data.info()
```

## convert category feature

```python
data['Sex'].value_counts()
```

```python
data['Embarked'].value_counts()
```

```python
sex=pd.get_dummies(data['Sex'],drop_first=True)
sex
```

```python
embark=pd.get_dummies(data['Embarked'])
embark
```

```python
# to merge new column in my dataset
data=pd.concat([data,sex,embark],axis=1)
```

```python
data
```

```python
# to drop unimportant column
data.drop(['PassengerId','Name','Sex','Ticket','Embarked'],axis=1,inplace=True)
data
```

```python
data.rename(columns={'male':'Gender'},inplace=True)
data
```
